```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# balanced, pre-1920
labels = 'roles/occupations', 'presence', 'negative'
sum = 1170 + 2615 + 92
sizes = [1170/sum, 2615/sum, 92/sum]

fix1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
plt.savefig('balanced-pre-1920-graph.png')
```

```python
# balanced, post-1920
labels = 'roles/occupations', 'presence', 'negative'
sum = 1007 + 4041 + 98
sizes = [1007/sum, 4041/sum, 98/sum]

fix1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
plt.savefig('balanced-post-1920-graph.png')
```

```python
# unbalanced, pre-1920
labels = 'roles/occupations', 'presence', 'negative'
sum = 8132 + 21479 + 893
sizes = [8132/sum, 21479/sum, 893/sum]

fix1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
plt.savefig('unbalanced-pre-1920-graph.png')
```
